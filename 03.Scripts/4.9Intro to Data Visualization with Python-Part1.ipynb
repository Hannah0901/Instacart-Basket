# convert any mixed-type data.

```python
df_customer.describe()
```

```python
# Check data types
df_customer.dtypes
```

```python
#change userid from int to string
df_customer['user_id'] = df_customer['user_id'].astype('str')
```

```python
# check for columns that have mixed data types that may make analysis in the future complicated
for col in df_customer.columns.tolist():
    weird = (df_customer[[col]].applymap(type) != df_customer[[col]].iloc[0].apply(type)).any(axis=1)
    if len (df_customer[weird]) > 0:
        print (col)
```

The first name column may be considered mixed since it contains NaN values, we can assign the data type as string

```python
# assign First Name as string
df_customer['First Name'] = df_customer['First Name'].astype('str')
```
